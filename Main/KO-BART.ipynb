# Kobart tutorial

```python
from transformers import PreTrainedTokenizerFast, BartModel, BartForConditionalGeneration
import torch
```

```python
tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-summarization')
model = BartForConditionalGeneration.from_pretrained('gogamza/kobart-summarization')
```

```python
text = """
(강릉=연합뉴스) 유형재 기자 = 강원 강릉시가 '아이 키우기 좋은, 아이 행복한 도시 조성'을 위해 올해 육아 지원 정책 확대와 돌봄 인프라 구축에 박차를 가한다.
우선 어린이집 야간연장 보육을 강화해 맞벌이 부모 근무로 인한 아동 보육 공백이 발생할 경우, 주간에 다른 어린이집을 이용하는 아동이더라도 야간연장 어린이집과 휴일 보육이 지정된 어린이집을 필요한 만큼 이용할 수 있도록 했다.
시는 2022년 10개소였던 야간연장 어린이집을 올해 33개소까지 확대한다.
일시적인 돌봄이 필요한 경우에는 7곳의 시간제 보육 어린이집을 이용할 수 있다.
2022년부터 강릉시 자체로 추진하는 '아이돌봄서비스료 이용자부담금 추가 지원'도 소득유형별 40∼80%로 확대된다.
초등 돌봄 사각지대 해소를 위해 기존 3개소 운영하던 '다함께돌봄센터'는 오는 9월부터 총 4개소를 운영한다.
올해부터는 월 이용료 6만원을 시에서 지원해 무료로 이용할 수 있으며, 시는 추가로 신규 시설을 확충해 나갈 예정이다.
작년 11월 유천동 복합복지체육센터 1층에 개관한 강릉시 최초의 '공공형 실내놀이터'는 아동 1인당 이용요금 2천원으로 아이와 부모의 큰 호응을 얻고 있다.
아울러 보육환경이 취약한 북부권 돌봄 사각지대를 해소하고자 올해 주문진 문화교육센터 2층에 장난감도서관이 신규 설치된다.
특히 24시간 운영하는 아동학대 긴급신고 전화의 건수가 매년 증가함에 따라 신속한 대응과 안전망 강화를 위해 올해 아동학대 전담 공무원을 2명에서 3명으로 증원한다.
학대 피해 아동의 안전한 보호와 치료 및 양육 서비스를 위한 '남아 전용 쉼터'도 신규 설치하여 7월부터 운영할 계획이다.
이 밖에도 결식아동 급식 지원, 보육교사 처우개선 지원, 평가제 어린이집 냉난방비 지원, 취약계층 아동 및 가구별 맞춤 지원 등 빈틈없는 보육환경 조성에 앞장선다.
김홍규 시장은 26일 "부모의 양육 부담을 줄이고 아동들에게 안전하고 유익한 서비스를 제공하기 위한 다양한 육아 지원 정책을 발굴하고 인프라를 확충해 나가겠다"고 말했다.
"""

text = text.replace('\n', ' ')

raw_input_ids = tokenizer.encode(text)
input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]

summary_ids = model.generate(torch.tensor([input_ids]),  num_beams=6,  min_length=24, max_length=256, bos_token_id=0, eos_token_id=1)
tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)
```
